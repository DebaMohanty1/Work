# Scikit-learn Transformation

### Scikit-learn has 3 types of transformation:
- Function Transformation
- Power Transformation
- Quantile transformation

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
from sklearn.preprocessing import FunctionTransformer
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import QuantileTransformer
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
X, y = fetch_california_housing(return_X_y=True, as_frame=True)
```

```python
X.head()
```

```python
y.head()
```

```python
X = X.iloc[:, :-2]
```

```python
X.head()
```

```python
X.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2, random_state=0)
```

```python
def plots(df, var, t):
    plt.figure(figsize=(13,5))
    plt.subplot(121)
    sns.kdeplot(df[var])
    plt.title('before ' + str(t).split('(')[0])
    
    plt.subplot(122)
    p1 = t.fit_transform(df[var])
    sns.kdeplot(p1)
    plt.title('after ' + str(t).split('(')[0])
```

### Function Transformation
- Based on function object or some user defined function
- Examples are Log, square-root etc.

```python
# Square Root Transformation
for col in X_train.columns:
    plots(X_train, col, FunctionTransformer(lambda x: x**(1/2)))
```

```python
# Cube Root Transformation
for col in X_train.columns:
    plots(X_train, col, FunctionTransformer(lambda x: np.power(x, (1/3))))
```

```python
# Log Transformation
for col in X_train.columns:
    plots(X_train, col, FunctionTransformer(np.log1p))
```

**After applying these transformation we can see a slightly better results than the real data, but it is not that significant if we consider normal distribution**

```python
# Square Root Transformation
f = FunctionTransformer(lambda x: np.power(x, 1/2))
```

```python
X_train['HouseAge'].head()
```

```python
f.fit(X_train['HouseAge'])
f.transform(X_train[['HouseAge']]).head()
```

```python
# Log Transformation
def model_accuracy_scaled(mod):
    model_scaled = Pipeline([
        ('scale', FunctionTransformer(np.log1p)),
        ('model', mod)
    ])
    model_scaled.fit(X_train, y_train)
    return model_scaled.score(X_test, y_test)

def model_accuracy_unscaled(mod):
    model_unscaled = Pipeline([
        ('model', mod)
    ])
    model_unscaled.fit(X_train, y_train)
    return model_unscaled.score(X_test, y_test)
```

```python
model_accuracy_scaled(KNeighborsRegressor())
```

```python
model_accuracy_unscaled(KNeighborsRegressor())
```

```python
model_accuracy_scaled(RandomForestRegressor(random_state=0))
```

```python
model_accuracy_unscaled(RandomForestRegressor(random_state=0))
```

**Here we can see a clear increase in the model accuracy in scaled data than in unscaled data.**<br>
but the decision tree model is unaffected by the transformation.

### Power Transformation
- Used when requirement is a Gaussian output
- Useful when dealing with heteroscedasticity(non constant variance)
- It has 2 methods
    - Box-Cox method
    - Yeo-Johnson method

```python
def plots(df, var, t):
    plt.figure(figsize=(13,5))
    plt.subplot(121)
    sns.kdeplot(df[var])
    plt.title('before ' + str(t).split('(')[0])
    
    plt.subplot(122)
    p1 = t.fit_transform(df[[var]]).flatten()
    sns.kdeplot(p1)
    plt.title('after ' + str(t).split('(')[0])
```

```python
# Box-Cox Transformation
for col in X.columns:
    plots(X, col, PowerTransformer(method='box-cox'))
```

```python
# Yeo-Johnson Transformation
for col in X.columns:
    plots(X, col, PowerTransformer())
```

#### Box-Cox Transformation
- Can only be used for positive values

```python
p = PowerTransformer(method='box-cox')
p.fit_transform(X_train)
```

```python
model_accuracy_scaled(KNeighborsRegressor())
```

```python
model_accuracy_unscaled(KNeighborsRegressor())
```

```python
model_accuracy_scaled(RandomForestRegressor(random_state=0))
```

```python
model_accuracy_unscaled(RandomForestRegressor(random_state=0))
```

### Quantile Transformation

```python
for col in X.columns:
    plots(X, col, QuantileTransformer(output_distribution='normal'))
```

```python
p = QuantileTransformer(output_distribution='normal')
p.fit_transform(X_train)
```

```python
p.quantiles_
```

```python
model_accuracy_unscaled(RandomForestRegressor(random_state=0))
```

From all of the above Transformation we can see that for KNN only these transformation is working as the KNN algorithm require a normal data. Random forest is unaffected the the data distribution, so we dont see a major change in the accuracy of scaled and unscaled data.
